```python
import pandas as pd

# train 데이터 불러오기
data = pd.read_csv('/data/ephemeral/home/data/train.csv')

data
```

```python
# id 컬럼 기준 중복 데이터 확인
id_duplicates = data[data.duplicated(subset='id', keep=False)]

id_duplicates
```

```python
# question + paragraph 기준 중복 데이터 확인
import ast

# 문자열로 저장된 딕셔너리를 실제 딕셔너리로 변환
data['problems'] = data['problems'].apply(ast.literal_eval)

# question + paragraph 조합 생성
data['question_paragraph'] = data.apply(lambda x: x['problems']['question'] + x['paragraph'], axis=1)

# 중복된 question + paragraph 조합 확인
qp_duplicates = data[data.duplicated(subset='question_paragraph', keep=False)]

qp_duplicates
```
